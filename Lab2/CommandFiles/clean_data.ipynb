#### 1.Wczytano plik csv jako df  

```python
import pandas as pd
df = pd.read_csv("../OriginalData/tb.csv")
df.head()
```

#### 2.przekształcono kolumny new_spm04 itd. na wiersze (melting)

```python


df_m = pd.melt(df, ["iso2", "year"])
df_m.head(10000)

```

#### 3.Rozbito informacje o płci i wieku na dwie kolumny(sex, age)

```python
df_m['sex'] = df_m['variable'].str[7]
df_m['age'] = df_m['variable'].str[8:]
df_m.head(10000)
```

#### 4. Usunięto wiersze dla których w kolumnie age nie było żadnych wartości

```python
index_names = df_m[df_m['age'].str.len() < 1].index
print(index_names)
df_m.drop(index_names, inplace = True)
df_m.head(10000)
```

#### 5. Zmieniono formatowanie tekstu w kolumnie age

```python
df_m['age'] = df_m['age'].apply(lambda x: x[0] + '-' + x[1:] if len(x) < 4 else x[0:2] + '-' + x[2:])
df_m.head(10000)
```

#### 6. Usunięto kolumnę w której dotychczas znajdowały się razem informacje o płci i wieku

```python
df_m.drop('variable', axis = 'columns', inplace = True)
df_m.head(10000)
```

#### 7.Przestawiono kolumnę cases na ostatnie miejsce w tabeli
#### 8. Usunięto wiersze dla których, nie było wartości w kolumnie cases

```python
last_col = df_m.pop('value')
df_m.insert(4,'cases', last_col)
df_m.dropna(axis=0, how='any', subset = ['cases'], inplace = True)
df_m.head(1000)
```
