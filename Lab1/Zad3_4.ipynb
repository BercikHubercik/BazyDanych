#### Zdefiniowano funkcję $f(x)=x^2+5$ 


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


def f(x):
    return  x**2 + 5

```

#### Zdefiniowano przedziały argumentów

```python
x1 = np.linspace(-1, 1, endpoint = False)
x2 = np.linspace(-6, 6, endpoint = False)
x3 = np.linspace(0, 5, endpoint = False)



```

#### Utworzono wykresy funkcji dla podanych przedziałów

```python
plt.plot(x1, f(x1))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Plot of function y = x^2 + 5")
plt.legend("y")
```

```python
plt.plot(x2, f(x2))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Plot of function y = x^2 + 5")
plt.legend("y")
```

```python
plt.plot(x3, f(x3))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Plot of function y = x^2 + 5")
plt.legend("y")
```
